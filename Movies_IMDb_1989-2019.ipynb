#### Reordenar las columnas del dataset:

```python
movies_imdb = movies_imdb[['Title','Director','Star','Genre', 'Certificate', 'Release_Date', 'Runtime', 'Country', 'Language', 'IMDb_Rating', 'IMDb_Votes', 'Production_Co', 'Gross', 'Budget']]
movies_imdb.head()
```

#### Cambiar el tipo de dato a la columna IMDb_votes para que se aplique el formato de miles:

```python
movies_imdb['IMDb_Votes'] = movies_imdb['IMDb_Votes'].astype(float)
movies_imdb.head()
```

#### Quitar el año a la columna Title:

```python
movies_imdb[['Title', '0', '1']] = movies_imdb.Title.str.split("(",expand=True,)
movies_imdb.head()
```

#### Quitar  espacios  en blanco a la columna Certificate:

```python
movies_imdb['Certificate']=movies_imdb['Certificate'].str.strip()
movies_imdb.head()
```

#### Quitar el dato del año a la columna de Release_Date:

```python
movies_imdb[['Release_Date','2']] = movies_imdb.Release_Date.str.split("(",expand=True,)
movies_imdb.head()
```

#### Quitar espacios en blanco de la columna Release_Date:

```python
movies_imdb['Release_Date']=movies_imdb['Release_Date'].str.strip()
movies_imdb.head()
```

```python
movies_imdb[['Day','Release_Month', 'Release_Year']] = movies_imdb.Release_Date.str.split(" ",expand=True,)
movies_imdb.head()
```

#### Dar formato de fecha a la columna Release_Date:

```python
movies_imdb['Release_Date'] =  pd.to_datetime(movies_imdb['Release_Date'],
                              format= '%d %B %Y')
movies_imdb.head()
```

#### Eliminar las columnas "basura", resultantes de los cambios anteriores

```python
movies_imdb=movies_imdb.drop(['0', '1', '2'], axis=1)
movies_imdb.head()
```

```python
movies_imdb['Release_Year'] = movies_imdb['Release_Year'].astype(int)
#movies_imdb.head()
```

```python
Years_ranges = ["{0} - {1}".format(year, year + 9) for year in range(1990, 2020, 10)]
Years_ranges
```

```python
count_unique_Years_ranges = len(Years_ranges)
count_unique_Years_ranges
```

```python
movies_imdb['Years_ranges'] = pd.cut(x=movies_imdb['Release_Year'], bins=count_unique_Years_ranges, labels=Years_ranges)
movies_imdb
```

```python
movies_imdb['IMDb_Rating'] = movies_imdb['IMDb_Rating']*10

def category_movie(x):
    if x <=39:
        return "Bad"
    if 40 < x <=69:
        return "Regular"
    else:
        return "Good"

#for col in movies_imdb.column:
movies_imdb['IMDb_CatRating'] = movies_imdb['IMDb_Rating'].apply(lambda x: category_movie(x))
movies_imdb['IMDb_CatRating'] = movies_imdb['IMDb_CatRating'].astype('category')
movies_imdb.head()
```

```python
print(type(pd.DatetimeIndex(movies_imdb['Release_Date']).month))
movies_imdb['Release_Month'] = pd.DatetimeIndex(movies_imdb['Release_Date']).month
movies_imdb.head()
```

```python
def category_month(x):
    if x <=3:
        return "I-Quater"
    if 3 < x <=6:
        return "II-Quater"
    if 6 < x <=9:
        return "III-Quater"
    else:
        return "IV-Quater"

#for col in movies_imdb.column:
movies_imdb['Release_Month'] = movies_imdb['Release_Month'].apply(lambda x: category_month(x))
movies_imdb['Release_Month'] = movies_imdb['Release_Month'].astype('category')
movies_imdb.head()
```

```python
movies_imdb['Budget_Category'] = movies_imdb['Budget']
bins = pd.IntervalIndex.from_tuples([(60000, 44250000), (44250001,88000000), (88000001, 321000000)])
movies_imdb['Budget_Category'] = pd.cut(movies_imdb['Budget'],bins )
movies_imdb.head()
```

```python
def category_budget(x):
    if 60000 <= x <= 20750000:
        return "Low Budget"
    if 44250001 <= x <= 88000000:
        return "Average Budget"
    else:
        return "High Budget"

#for col in movies_imdb.column:
movies_imdb['Budget_Category'] = movies_imdb['Budget'].apply(lambda x: category_budget(x))
movies_imdb['Budget_Category'] = movies_imdb['Budget_Category'].astype('category')
movies_imdb.head()
```
