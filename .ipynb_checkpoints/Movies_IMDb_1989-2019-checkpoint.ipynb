# Movies IMDb 1980 - 2019

```python
import pandas as pd
import numpy as np
import datetime

#Aplicar formato de miles y decimales a los datos 
pd.options.display.float_format = '{:,.2f}'.format
```

#### Importar dataset e inspeccionar las primeras lineas del mismo:

```python
movies_imdb = pd.read_csv("Titles.csv", na_values=['0'])
movies_imdb.head()
```

#### Para empezar a trabajar columna a columna se decide reordenar las columnas del dataset:

```python
movies_imdb = movies_imdb[['Title','Director','Star','Genre', 'Certificate', 'Release_Date', 'Runtime', 'Country', 'Language', 'IMDb_Rating', 'IMDb_Votes', 'Production_Co', 'Budget', 'Gross']]
movies_imdb.head()
```

#### Inspeccionar el tipo de dato de cada columna y la cantidad de datos presentes en el dataset: 

```python
movies_imdb.info()
```

#### Cambiar el tipo de dato a la columna IMDb_votes para que se aplique el formato de miles:

```python
movies_imdb['IMDb_Votes'] = movies_imdb['IMDb_Votes'].astype(float)
movies_imdb.head()
```

#### Quitar el año a la columna Title:

```python
movies_imdb[['Title', '0', '1']] = movies_imdb.Title.str.split("(",expand=True,)
movies_imdb.head()
```

#### Quitar  espacios  en blanco a la columna Certificate:

```python
movies_imdb['Certificate']=movies_imdb['Certificate'].str.strip()
movies_imdb.head()
```

#### En la columna Certificate contamos con varias categorías que pueden consolidarse bajo la misma "Not Rated":

```python
movies_imdb['Certificate'].value_counts()
```

#### Reemplazar las categorías "Unrated" bajo el nombre "Not Rated":

```python
movies_imdb['Certificate'] = movies_imdb['Certificate'].replace({'Unrated': 'Not Rated', 'nan': 'Not Rated', ' ': 'Not Rated'})
movies_imdb.head()
```

#### Quitar el dato del país a la columna de Release_Date:

```python
movies_imdb[['Release_Date','2']] = movies_imdb.Release_Date.str.split("(",expand=True,)
movies_imdb.head()
```

#### Quitar espacios en blanco de la columna Release_Date:

```python
movies_imdb['Release_Date']=movies_imdb['Release_Date'].str.strip()
movies_imdb.head()
```

#### Segmentar Release_Date para obtener el mes y el año de lanzamiento de la película

```python
movies_imdb[['Day','Release_Month', 'Release_Year']] = movies_imdb.Release_Date.str.split(" ",expand=True,)
movies_imdb.head()
```

#### Dar formato de fecha a la columna Release_Date:

```python
movies_imdb['Release_Date'] =  pd.to_datetime(movies_imdb['Release_Date'],
                              format= '%d %B %Y')
movies_imdb.head()
```

#### Quitar la palabra min que acompaña a los datos en la columna Runtime:

```python
movies_imdb[['Runtime','Min.']] = movies_imdb.Runtime.str.split(" ",expand=True,)
movies_imdb.head()
```

#### Convertir el Runtime a formato numérico para poder trabajar los valores faltantes:

```python
movies_imdb['Runtime']=movies_imdb['Runtime'].astype(float)
```

#### Completar los valores faltantes de la columna Runtime con el promedio de esta columna:

```python
movies_imdb['Runtime'] = movies_imdb['Runtime'].fillna((movies_imdb['Runtime'].mean()))
movies_imdb.head()
```

#### Eliminar las columnas "basura", resultantes de los cambios anteriores

```python
movies_imdb=movies_imdb.drop(['0', '1', '2', 'Day', 'Min.'], axis=1)
movies_imdb.head()
```

#### Se establecen categorías por rangos de rating de las películas:

```python
#cambio de la escala de IMDb_Rating
movies_imdb['IMDb_Rating'] = movies_imdb['IMDb_Rating']*10

#establecimiento de las categorías por rango
def category_movie(x):
    if x <=39:
        return "Bad"
    if 40 < x <=69:
        return "Regular"
    else:
        return "Good"

#creación de la columna
movies_imdb['Rating_Category'] = movies_imdb['IMDb_Rating'].apply(lambda x: category_movie(x))
movies_imdb['Rating_Category'] = movies_imdb['Rating_Category'].astype('category')
movies_imdb.head()
```

#### Pasar los meses (nombres) al número de mes correspondiente:

```python
movies_imdb['Release_Month'] = pd.DatetimeIndex(movies_imdb['Release_Date']).month
movies_imdb.head()
```

#### Establecer rangos por trimestre de lanzamiento de las películas:

```python
#establecimiento de las categorías por rango
def category_month(x):
    if x <=3:
        return "Q1"
    if 3 < x <=6:
        return "Q2"
    if 6 < x <=9:
        return "Q3"
    else:
        return "Q4"

#creación de la columna
movies_imdb['Release_Month'] = movies_imdb['Release_Month'].apply(lambda x: category_month(x))
movies_imdb['Release_Month'] = movies_imdb['Release_Month'].astype('category')

#renombrar columna
movies_imdb=movies_imdb.rename(columns = {'Release_Month':'Release_Period'})
movies_imdb.head()
```

#### Cambiar el tipo de dato de Release_Year:

```python
movies_imdb['Release_Year'] = movies_imdb['Release_Year'].astype(int)
```

#### Establecer décadas de lanzamiento de las películas con ayuda d ela función range():

```python
Years_by_decades = ["{0}s".format(year) for year in range(1980, 2020, 10)]
Years_by_decades
```

#### Guardar el dato de largo de la variable del punto anterior:

```python
count_unique_Years_by_decades = len(Years_by_decades)
count_unique_Years_by_decades
```

#### Mediante la función .cut() crear las décadas y crear la nueva columna Release_Decade en el dataframe:

```python
movies_imdb['Release_Decade'] = pd.cut(x=movies_imdb['Release_Year'], bins=count_unique_Years_by_decades, labels=Years_by_decades)
movies_imdb.head()
```

#### Total de peliculas por década:

```python
movies_imdb['Release_Decade'].value_counts()
```

#### Agrupamos los datos promedio de Budget y Gross por Género de película:

```python
mean_by_genre= movies_imdb.groupby('Genre')[['Budget','Gross']].mean()
mean_by_genre
```

#### El siguiente código rellena los valores nas de las columnas Budget y Gross con el promedio del monto del género de cada columna:

```python
movies_imdb.fillna(movies_imdb.groupby('Genre')[['Budget', 'Gross']].transform('mean'), inplace=True)
movies_imdb.head()
```

#### Para considerar los valores en la columna Budget como válidos para nuestro posterior análisis, se han conseguido los índices de conversión del dinero en el tiempo de Estados Unidos.

#### Se importa un .csv con los índices promedio por año:

```python
index= pd.read_csv("Adjust_Index.csv")
```

#### Con ayuda de la función merge() integramos los datos de ambos dataframes movies_imdb e index:

```python
movies_imdb = pd.merge(movies_imdb, index)
movies_imdb.head()
```

#### Se crea una nueva columna que guarde el dato del Budget ajustado:

```python
movies_imdb['Adjusted_Budget']= (movies_imdb['Budget']*movies_imdb['Index'])
movies_imdb.head()
```

```python
movies_imdb.describe()
```

#### Se establacen rangos basados en los cuartiles de la columna para los montos de Adjusted_Budget requeridos para las películas:

```python
#creación de una nueva columna con los valores de la columna Budget
movies_imdb['Budget_Category'] = movies_imdb['Adjusted_Budget']

#creación de los bins
bins = pd.IntervalIndex.from_tuples([(9000, 27678000), (27678001,96853000), (96853001, 365000000)])

#dividir la columna en los bins indicados 
movies_imdb['Budget_Category'] = pd.cut(movies_imdb['Adjusted_Budget'],bins )
movies_imdb.head()
```

#### Los rangos creados arriba se reescribirán con el nombre de la categoría:

```python
#establecimiento de las categorías por rango
def category_budget(x):
    if 90000 <= x <= 27678000:
        return "Low Budget"
    if 27678001 <= x <= 96853000:
        return "Average Budget"
    else:
        return "High Budget"

#creación de la columna
movies_imdb['Budget_Category'] = movies_imdb['Budget'].apply(lambda x: category_budget(x))
movies_imdb['Budget_Category'] = movies_imdb['Budget_Category'].astype('category')
movies_imdb.head()
```

#### Cantidad de películas por rango:

```python
movies_imdb['Budget_Category'].value_counts()
```

#### Se decide filtrar los valores nulos de Titles.csv en base a la columna Gross:

```python
movies_imdb = movies_imdb.dropna(subset=['Gross'])
movies_imdb.head()
```

#### Como resultado se obtiene un dataframe de 1992 filas con sus datos completos en cada una de las columnas:


```python
movies_imdb.info()
```

#### Reordenar columnas:

```python
movies_imdb = movies_imdb[['Title','Director','Star','Genre', 'Certificate', 'Release_Date', 'Release_Period', 'Release_Year', 'Release_Decade', 'Runtime', 'Country', 'Language', 'IMDb_Rating', 'Rating_Category', 'IMDb_Votes', 'Production_Co', 'Budget', 'Index', 'Adjusted_Budget','Budget_Category', 'Gross']]
movies_imdb.head()
```

#### Exportar a .csv para trabajar visualización en la entrega final del proyecto:

```python
movies_imdb.to_csv('Titles_IMDb.csv')
```
